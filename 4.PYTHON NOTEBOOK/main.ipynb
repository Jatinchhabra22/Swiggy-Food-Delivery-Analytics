```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('/Users/jatinchhabra/Desktop/SWIGGY ANALYSIS/2.DATASETS/swiggy.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.rename(columns={
    'Avg ratings' : 'avg_ratings',
    'Total ratings': 'total_ratings',
    'Food type': 'food_type',
    'Delivery time': 'delivery_time',
    'Price': 'price'                          
},inplace=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
#### PHASE 2 --- FEATURE ENGINEERING 

## 1) PRICE RANGE 

def price_range(x):
    if x < 200:
        return "budget"
    elif x <= 400:
        return "mid range"
    else:
        return "premium"
    
dataset['Price_range'] = dataset['price'].apply(price_range)
```

```python
dataset.head()
```

```python
dataset['avg_ratings'].unique()
```

```python
## 2) ratings feature 

def rating(r):
    if r <= 3:
        return "below avg rated"
    elif r > 3 and r <= 4:
        return "avg rated"
    else:
        return "High rated"
    
dataset['Rating_range'] = dataset['avg_ratings'].apply(rating)
```

```python
dataset.head()
```

```python
dataset['delivery_time'].unique()
```

```python
## 3) Delivery speed 

def speed(t):
    if t <= 40:
        return "Fast Delivery"
    elif t <= 55:
        return "Moderate"
    else:
        return "slow "
    
dataset['Delivery_range'] = dataset['delivery_time'].apply(speed)
```

```python
dataset.head()
```

```python
## 4) Primary cusiness

# primary cuisine as first one
dataset['primary_cuisine'] = dataset['food_type'].str.split(',').str[0].str.strip()
```

```python
dataset.head()
```

```python
### 6) number of cuisines offered
dataset['num_of_cuisines'] = dataset['food_type'].str.split(',').apply(len)
```

```python
dataset.head()
```

```python
### 7) popularity score --- no of total ratings * rating

dataset['popularity_scores'] = dataset['avg_ratings'] * np.log1p(dataset['total_ratings'])
```

```python
dataset.head()
```

```python
### PHASE 3 --- GIVING AND ANALYSING STRONG PREDICTIONS BASED ON OUR DATASET

### TOP 10 RESTRAUNTS BY POP_SCORE

top_restraunts = dataset.sort_values(['popularity_scores'],ascending=False).head(10)
```

```python
top_restraunts
```

```python
## 2) best value for money restaurants ( less price , high rating)

best_value = dataset[dataset['avg_ratings'] >= 4.0].sort_values('popularity_scores', ascending=False).head(10)
```

```python
best_value
```

```python
dataset['popularity_scores'].unique()
```

```python
def pop_bracet(score):
    if score > 25.000:
        return "High Popularity"
    elif score > 10 and score < 24.99:
        return "Medium Popularity"
    else:
        return "Less popularity"

dataset['Pop_bracket'] = dataset['popularity_scores'].apply(pop_bracet)
```
